# AÃ±adir dÃ­a de la semana

```python
# Creo indices de sobra. Pongo que empiece en -2 porque el dÃ­a 0 es miercoles
indices_dias = np.arange(-2,103).reshape(105,1)

# DÃ­a de la semana
dias_semana = ['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo']
dias = np.array(dias_semana*15).reshape(105,1)

# Construccion de la semana
dias_semanas = pd.DataFrame([i[0],j[0]] for i,j in zip(indices_dias,dias) )
dias_semanas.columns = ['date_number','Dia_semana']
dias_semanas = dias_semanas.loc[dias_semanas.date_number >= 0 ].reset_index(drop=True)
```

```python
train = train.join(dias_semanas.set_index('date_number'), on='date_number')
```
